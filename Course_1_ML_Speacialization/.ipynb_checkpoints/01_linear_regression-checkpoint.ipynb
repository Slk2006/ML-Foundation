## 1D implementation

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(7)

m = 10 #NO. of Data points
x = np.linspace(0, 10, m)

noise = np.random.randn(m) * 2 #Gaussian noise
y = 2 * x + noise

plt.scatter(x, y)
plt.title("Training Data: Y = 2X + noise")
plt.xlabel("X - Feature 1")
plt.ylabel("Y - Label")
plt.show()
```

#### Initialise Parameters and Predict Function 

```python
# Parameters - b = intercept, w = slope

w = 0.0
b = 0.0

def predict(w, x, b):
    return ((w * x) + b)
```

#### Define Cost Function MSE

```python
def compute_cost(x, y, w, b):
    y_pred = predict(w, x, b)
    cost = (1/(2* m )) * np.sum((y_pred - y) ** 2)
    return cost

print("Initial cost:", compute_cost(x, y, w, b))
```

#### GRADIANT DESCENT

```python
# Learning Rate
a = 0.01
# iterations
iter = 100

cost_hstry = []
for i in range(iter):
    y_pred = predict(w, x, b)

    # Gradient
    d_w = (1/ m) * (np.sum((y_pred - y) * x))
    d_b = (1/ m) * (np.sum(y_pred - y))

    # updates
    
    w -= a * d_w
    b -= a * d_b

    # Cost for new parameters
    cost = compute_cost(x, y, w, b)
    cost_hstry.append(cost)

print("Final W:", w)
print("Final B:", b)
print("Final Cost:", cost_hstry[-1])
```

```python
plt.plot(cost_hstry)
plt.xlabel("Iter")
plt.ylabel("COST")
plt.title("Iteration v/s COST - Gradiant Descent")
plt.show()
```

```python
plt.scatter(x, y, label = "Data")
y_fit = predict(w, x, b)
plt.plot(x, y_fit, color = "red", label = "Fitted line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```

## Multivariate Linear Regression

```python
# lets create a dataset with 2 features
# X1 is size and X2 is No. of rooms

np.random.seed(56)

m = 20
# size of house in sq. ft
x1 = np.random.uniform(500, 4000, m)
# No of Rooms
x2 = np.random.randint(1, 6, m)
noise = np.random.randn(m) * 20000
# True relationship 
y = 50 * x1 + 10000 * x2 + noise + 50000

X = np.column_stack((x1, x2))
X
```

#### FEATURE SCALING

```python
# Mean Normalisation
mu = np.mean(X, axis = 0)
sigma = np.std(X, axis = 0)

X_normal = (X - mu) / sigma

# Add Bias colums of ones
X_bias = np.c_[np.ones((m, 1)), X_normal] #shape = (m, 3)
X_bias
```

```python
W = np.zeros(3, )

def compute_cost_vec(X, y, W):
    y_pred = X.dot(W)
    cost = (1/(2 * m)) * np.sum((y_pred - y) ** 2)
    return cost

def grad_descent_vect(X, y, W, a, iter):
    cost_hstry = []

    for i in range(iter):
        y_pred = X.dot(W)
        gradient = (1/m) * X.T.dot(y_pred - y)
        W = W - a * gradient
        cost_hstry.append(compute_cost_vec(X, y, W))
        return W, cost_hstry

a = 0.01
iter = 1000

W_final, cost_hstry = grad_descent_vect(X_bias, y, W, a, iter)

print("Final W: ", W_final)
print("Fianl Cost: ", cost_hstry[-1])
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_bias, y)
print("Sklearn intercept:", lin_reg.intercept_)
print("Sklearn coefficients:", lin_reg.coef_)
```
