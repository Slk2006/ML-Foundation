## # 03 - Overfitting and Regularization

### Polynomial regression on synthetic data to see overfitting and the effect of L2 regularization.


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error as MSE

%matplotlib inline
```

```python
np.random.seed(9)

m = 60
X = np.linspace(-3, 3, m).reshape(-1, 1)
# true function
y_true = .5 * (X[:, 0] ** 2)
noise = np.random.randn(m) * 1.5
y = y_true + noise

plt.scatter(X, y, s = 15)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Data for Polynomial")
plt.show()
```

```python
m_train = 40
X_train, y_train = X[:m_train], y[:m_train]
X_val, y_val = X[m_train:], y[m_train:]
```

```python
def fit_plot_poly(degree, a = 0.0, use_ridge = False):
    poly = PolynomialFeatures(degree= degree, include_bias = False)
    X_train_poly = poly.fit_transform(X_train)
    X_val_poly = poly.transform(X_val)
    X_all_poly = poly.transform(X)

    if use_ridge and a > 0:
        model = Ridge(alpha = a)
        title_extra = f"Ridge (lamda = {a})"
    else:
        model = LinearRegression()
        title_extra = 'No Regularization'

    model.fit(X_train_poly, y_train)

    y_train_pred = model.predict(X_train_poly)
    y_val_pred = model.predict(X_val_poly)
    y_all_pred = model.predict(X_all_poly)

    train_mse = MSE(y_train, y_train_pred)
    val_mse = MSE(y_val, y_val_pred)

    print(f"Degree {degree}, {title_extra}")
    print(f"Train MSE: {train_mse:.3f}")
    print(f"  Val MSE: {val_mse:.3f}\n")

    # plot 
    plt.scatter(X_train, y_train, s=15, label = "Train")
    plt.scatter(X_val, y_val, s=15, label = "Val", alpha = 0.7)
    xs = np.linspace(-3, 3, 300).reshape(-1, 1)
    xs_poly = poly.transform(xs)
    ys_pred = model.predict(xs_poly)
    plt.plot(xs, ys_pred, color='red', label = f"Degree {degree}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Degree {degree} | {title_extra}")
    plt.legend()
    plt.show()
```

```python
for d in [1, 2, 10]:
    fit_plot_poly(degree=d)
```

```python
for lam in [0.0, 0.1, 1.0, 10.0, 100.0]:
    fit_plot_poly(degree=10, alpha=lam, use_ridge=True)
```
